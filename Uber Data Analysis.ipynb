```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

# Reading Database

```python
df = pd.read_csv('Uber-raw-data-sep14.csv', sep = ',')
```

```python
df.head(10)
```

#### Converting 'Date/Time' to panda recognizable 'Date/Time'

```python
df['Date/Time'] = df['Date/Time'].map(pd.to_datetime)
```

```python
df.head(3)
```

#### Getting Date Of Month , Weekday, Hour

```python
def get_Day_of_Month(df):
    return df.day
def get_Weekday(df):
    return df.weekday()
def get_Hour(df):
    return df.hour
```

```python
df['Day_Of_Month'] = df['Date/Time'].map(get_Day_of_Month)
df['Weekday'] = df['Date/Time'].map(get_Weekday)
df['Hour'] = df['Date/Time'].map(get_Hour)
```

```python
df.head(3)
```

# Performing Analysis

### Analysis For Date Of Month

```python
plt.hist(df['Day_Of_Month'], bins = 30, rwidth = 0.8, range= (0.5,30.5));
plt.grid()
plt.xlabel('Day Of Month')
plt.ylabel('Frequrncy')
plt.title('Frequency By Date Of Month')
```

### Analysis Of Weekday

```python
plt.hist(df['Weekday'], bins = 7, rwidth = 0.5, range = (-0.5, 6.5))
plt.grid()
plt.xticks(range(7), 'Mon Tue Wed Thu Fri Sat Sun'.split())
plt.xlabel('Weekday')
plt.ylabel('Frequency')
plt.title('Frequency By Weekday')
```

### Analysis Of Hour

```python
plt.hist(df['Hour'], bins = 24, range = (0,23))
plt.xlabel('Hour')
plt.ylabel('Frequecy')
plt.title('Frequency by Hour')
```

### Analysis On location

```python
plt.figure(figsize = (20,20))
plt.plot(df['Lon'], df['Lat'], '.', ms=0.105, alpha=.5)
plt.xlim(-74.2, -73.7)
plt.ylim(40.7, 41)
```
